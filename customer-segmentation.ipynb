```python
X=['0-30','31-60','61-90','91-120','120+']
Y=[len(df0_30.values),len(df31_60.values),len(df61_90.values),len(df91_120.values),len(df120above.values)]
plt.figure(figsize=(15,6))
sns.barplot(x=X,y=Y)
plt.xlabel('Annual Income (k$)')
plt.show()
```

```python
X1=df[['Age','Spending Score (1-100)']].values
from sklearn.cluster import KMeans
clusters=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(X1)
    clusters.append(kmeans.inertia_)
plt.figure(figsize=(15,6))
plt.plot(range(1,11),clusters,marker='o',color='red')
plt.show()
```

```python
kmeans=KMeans(n_clusters=4)
label=kmeans.fit_predict(X1)
```

```python
plt.scatter(X1[:,0],X1[:,1],c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='black')
plt.show()
```

```python
X2=df[['Annual Income (k$)','Spending Score (1-100)']].values
from sklearn.cluster import KMeans
clusters=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(X2)
    clusters.append(kmeans.inertia_)
plt.figure(figsize=(15,6))
plt.plot(range(1,11),clusters,marker='o',color='red')
plt.show()
```

```python
kmean=KMeans(n_clusters=5)
label=kmean.fit_predict(X2)
```

```python
plt.scatter(X2[:,0],X1[:,1],c=kmean.labels_,cmap="rainbow")
plt.scatter(kmean.cluster_centers_[:,0],kmean.cluster_centers_[:,1],color='black')
plt.show()
```
